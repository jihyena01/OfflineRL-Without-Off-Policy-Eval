```python
import torch

# 파일 경로
file_path = '/home/comet/Documents/KAIST/RL/onestep-rl/results/bullet-halfcheetah-medium-v0_beta/train_bullet-halfcheetah-medium-v0_0_reverse_kl_pybullet.pt.pt_10.pt'

# 파일 로드
data = torch.load(file_path)

# 데이터 타입과 내용 확인
print(type(data))
print(data)
```

```python
import matplotlib.pyplot as plt

# 데이터 준비
episodes = np.arange(1, len(data) + 1)
percentages = np.array(data) * 100

# 그래프 생성
plt.figure(figsize=(10, 6))
plt.plot(episodes, percentages, marker='o', linestyle='-', color='b')
plt.title('Agent Performance Over Episodes')
plt.xlabel('Episode')
plt.ylabel('Performance (%)')
plt.grid(True)
plt.show()
```
